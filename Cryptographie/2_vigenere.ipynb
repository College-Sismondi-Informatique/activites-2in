## B) Casser le chiffrement de Vigenère

Le chiffre de Vigenère est un **chiffre polyalphabétique**.  
Même s’il est plus robuste que le chiffre de César, il peut être cassé en utilisant des techniques de **cryptanalyse**, notamment l’**analyse de fréquence** et la recherche de motifs dans le texte.

Voici une méthode classique pour tenter de casser ce chiffrement.

### 1. Déterminer la longueur de la clé

La première étape consiste à **estimer la longueur de la clé utilisée pour le chiffrement**.

Pour cela, on peut utiliser **l’indice de coïncidence (IC)**.

L’indice de coïncidence mesure la probabilité que **deux lettres choisies au hasard dans le texte soient identiques**.

Il peut s’écrire sous la forme :

$$
IC = \sum_{i=1}^{26} p_i^2
$$

où \(p_i\) est la probabilité d'apparition de la lettre \(i\).

Pour un texte aléatoire (où toutes les lettres apparaissent avec la même probabilité), l’IC vaut environ :

$$
IC \approx 0.038
$$

Pour un texte en français, l’IC est généralement plus élevé :

$$
IC \approx 0.077
$$

#### Tester différentes longueurs de clé

On teste plusieurs longueurs possibles pour la clé (par exemple de 1 à 20).

Pour chaque longueur :

1. On divise le texte chiffré en plusieurs groupes.
2. On calcule l’indice de coïncidence pour chacun de ces groupes.
3. On compare les résultats avec l’IC attendu pour une langue naturelle.

Si l’IC obtenu est **proche de celui du français**, cela suggère que la longueur de clé testée est probablement correcte.

### 2. Diviser le texte chiffré

Une fois une **longueur de clé probable** trouvée, on divise le texte en plusieurs **sous-textes**.

Chaque sous-texte correspond à **une position de la clé**.

Par exemple, si la clé a une longueur de 3 :

- sous-texte 1 : lettres aux positions 0, 3, 6, 9, ...
- sous-texte 2 : lettres aux positions 1, 4, 7, 10, ...
- sous-texte 3 : lettres aux positions 2, 5, 8, 11, ...

Chaque sous-texte correspond alors à un **chiffrement de César**.

### 3. Analyser chaque sous-texte

On effectue ensuite une **analyse de fréquence** sur chaque sous-texte.

Dans un texte français, la lettre **E** est généralement la plus fréquente.

Si la lettre la plus fréquente dans un sous-texte est, par exemple, **K**, on peut supposer que **K correspond à E après décalage**.  
Cela permet d’estimer le **décalage utilisé**, donc la lettre correspondante de la clé.

### 4. Reconstituer la clé

En répétant ce processus pour chaque sous-texte, on peut déterminer **toutes les lettres de la clé**.

On obtient ainsi une **clé probable** utilisée pour le chiffrement.

### 5. Déchiffrer le texte

Une fois la clé trouvée, on peut **déchiffrer le texte** en appliquant l’opération inverse du chiffrement de Vigenère.

### Remarque

Cette méthode repose sur des **hypothèses statistiques**.  
Il est parfois nécessaire d’effectuer **plusieurs essais** pour trouver la bonne clé.

L’utilisation d’un **programme informatique** permet d’automatiser ces calculs et de tester rapidement plusieurs hypothèses.
