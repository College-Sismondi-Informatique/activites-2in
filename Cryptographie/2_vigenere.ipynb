## B) Casser le chiffrement 
Casser le chiffre de Vigenère, qui est un chiffre polyalphabétique, implique généralement l'utilisation de techniques d'analyse de fréquence et de recherche de motifs. Voici une méthode courante pour le casser :

<b>1. Déterminer la longueur de la clé :</b>
*    Calculer l'indice de coïncidence : L'indice de coïncidence (IC) est calculé en prenant le nombre de paires de lettres identiques dans le texte chiffré et en le divisant par le nombre total de lettres. Pour un texte chiffré avec une clé, l'IC sera plus élevé que pour un texte aléatoire.
$$
IC = \sum_{i=1}^{26} p_i^2
$$
où : $p_i$ est la probabilité d'apparition de la lettre $i$


*    Tester différentes longueurs de clé : Pour chaque longueur de clé possible (par exemple, de 1 à 20), divisez le texte chiffré en groupes de cette longueur et calculez l'IC pour chaque groupe. Comparez les résultats avec les valeurs typiques d'IC pour le texte en clair (par exemple, l'IC pour le français est d'environ 0,077).

*    Identifier la longueur de la clé : Une longueur de clé qui produit un IC plus élevé que les autres est probablement la bonne.




<b>2. Diviser le texte chiffré :</b><br>
Une fois que vous avez une estimation de la longueur de la clé, divisez le texte chiffré en plusieurs sous-textes, chacun correspondant à une position de la clé. Par exemple, si la clé a une longueur de 3, vous aurez trois sous-textes : un pour les lettres à des positions 0, 3, 6, etc., un pour les positions 1, 4, 7, etc., et un pour les positions 2, 5, 8, etc.

<b>3. Analyser chaque sous-texte :</b><br>
Pour chaque sous-texte, effectuez une analyse de fréquence pour déterminer quelle lettre est la plus fréquente. Dans un texte en français, la lettre 'E' est souvent la plus fréquente. En décalant cette lettre vers 'E', vous pouvez estimer le décalage utilisé pour cette position de la clé.

<b>4. Reconstituer la clé : </b><br>
Répétez le processus pour chaque sous-texte afin de reconstituer la clé complète.

<b>5. Déchiffrer le texte :</b><br>
Une fois que vous avez la clé, utilisez-la pour déchiffrer le texte chiffré en inversant le processus de chiffrement.

Ces étapes nécessitent souvent des essais et des erreurs, ainsi qu'une certaine intuition sur la langue du texte chiffré. Des outils informatiques peuvent également faciliter ce processus.
