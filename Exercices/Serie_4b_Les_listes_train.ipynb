Si notre horaire est construit ainsi :

````python
# Exemple de base (à adapter à votre 'horaire')
train1 = ['S22', '8h47', 'Vufflens-la-Ville', 1]
train2 = ['S2',  '8h51', 'Aigle',             4]
train3 = ['S4',  '8h55', 'Palézieux',         3]
train4 = ['S1',  '8h57', 'Lausanne',          4]

horaire = []

horaire.append(train1)
horaire.append(train2)
horaire.append(train3)
horaire.append(train4)

# Les 4 lignes précédantes peuvent être remplacées par une boucle for :
# la variable t prendra sucessivenemnt les valeurs des listes train1, train2, train3, train4.
#for t in (train1, train2, train3, train4):
#    horaire.append(t)
````

On peut également afficher l’entier de l’horaire dans la console sous une forme agréable à l’aide d’une boucle `for`:

```
for i in range(len(horaire)):
    print(horaire[i])
```

````python
# Exemple de base (à adapter à votre 'horaire')
train1 = ['S22', '8h47', 'Vufflens-la-Ville', 1]
train2 = ['S2',  '8h51', 'Aigle',             4]
train3 = ['S4',  '8h55', 'Palézieux',         3]
train4 = ['S1',  '8h57', 'Lausanne',          4]

horaire = []

for t in (train1, train2, train3, train4):
    horaire.append(t)

for i in range(len(horaire)):
    print(horaire[i])
````
